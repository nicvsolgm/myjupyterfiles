```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
quarter= ['1st','2nd','3rd','4th']
fstyr= [82,87,84,86]
sndyr = [90,86,86,88]
trdyr = [79,82,89,91]
frtyr = [85,86,88,92]
plt.plot(quarter,fstyr,"g*--")
plt.plot(quarter,sndyr,"ro--")
plt.plot(quarter,trdyr,"b*-.")
plt.plot(quarter,frtyr,"y+-.")
plt.title("students average grades per quarter finals ")
plt.xticks(["1st","2nd","3rd","4th"])
plt.yticks(np.arange(78,94,1))
plt.xlabel("quarter finals")
plt.ylabel("Student grades")
plt.legend(["first year","second year","third year","fourth year"])
plt.grid()
plt.savefig('Grades.png',dpi=200)
plt.figure(figsize=(30,10),dpi=100)
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 3., 0.1)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
