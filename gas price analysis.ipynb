```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns



```

```python
gas = pd.read_csv('gasprices2.csv')
gas.head()
```

```python
gas.shape[1]
```

```python
plt.plot(gas.Year,gas.USA, label = "USA")
plt.grid()
plt.xticks(gas.Year[::2])
plt.yticks(np.arange(0.90,3.5,0.15))
plt.legend()
plt.show()
```

```python
plt.plot(gas.Year,gas.Canada, label = "Canada")
plt.grid()
plt.xticks(gas.Year[::2])
plt.yticks(np.arange(0.90,4.5,0.25))
plt.legend()
plt.show()
```

```python
plt.title("Gas Price per Country", fontdict={'fontweight':'bold','fontsize':20})

plt.plot(gas.Year,gas.Australia,'y.-',label ='Australia')
plt.plot(gas.Year,gas.USA,'b.-',label ='United States')
plt.plot(gas.Year,gas.Canada,'r.-', label = 'Canada')
plt.plot(gas['Year'],gas['South Korea'],'g.-', label = 'South Korea')
plt.xticks(gas.Year[::2])
plt.yticks(np.arange(0,7,0.5))
plt.xlabel('Year',color='pink',fontdict={'fontweight':'bold','fontsize':17})
plt.ylabel('US Dollars',color = 'violet',fontdict={'fontweight':'bold','fontsize':16})
plt.legend()
plt.savefig('gas_price_figure.png', dpi = 300)
plt.grid()
plt.show()
```
