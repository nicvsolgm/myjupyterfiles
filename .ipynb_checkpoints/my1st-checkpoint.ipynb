```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns



```

```python
quarter= (['1st','2nd','3rd','4th'])
fstyr= ([82,87,84,86])
sndyr = ([90,86,86,88])
trdyr = ([79,82,89,91])
frtyr = ([85,86,88,92])
plt.plot(quarter,fstyr,"g*--")
plt.plot(quarter,sndyr,"ro--")
plt.plot(quarter,trdyr,"b*-.")
plt.plot(quarter,frtyr,"yo-.")
plt.title("students average grades per quarter finals ")
plt.xticks(["1st","2nd","3rd","4th"])
plt.yticks(np.arange(78,94,2))
plt.xlabel("quarter finals")
plt.ylabel(" average grades per quarter")
plt.legend(["first year","second year","third year","fourth year"])
plt.grid()
plt.figure(figsize=(20,6),dpi=100)
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 3., 0.1)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
plt.title("Gas Price per Country", fontdict={'fontweight':'bold','fontsize':20})

plt.plot(gas.Year,gas.USA,'b.-',label ='United States')
plt.plot(gas.Year,gas.Canada,'r.-', label = 'Canada')
plt.plot(gas['Year'],gas['South Korea'],'g.-', label = 'South Korea')
plt.xticks(gas.Year[::3])
plt.xlabel('Year',color='yellow',fontdict={'fontweight':'bold','fontsize':17})
plt.ylabel('US Dollars',color = 'violet',fontdict={'fontweight':'bold','fontsize':16})
plt.legend()
plt.savefig('gas_price_figure.png', dpi = 300)
plt.grid()
plt.show()
```

```python
for country in gas:
   if country!='Year':
    plt.plot(gas.Year,gas[country],marker='.')

plt.xticks(gas.Year[::2])
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.title("Gas Price per Country")
plt.grid()
plt.show()
```
